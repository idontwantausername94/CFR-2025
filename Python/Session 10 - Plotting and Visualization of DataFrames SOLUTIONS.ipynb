## Introduction to Plotting with Pandas

### What is Pandas Plotting?
Pandas integrates with Matplotlib to provide a high-level interface for plotting data contained in DataFrames and Series. This makes it easy to create quick and beautiful visualizations directly from Pandas objects.

### Importing Required Libraries
Before we start, let's import Pandas and Matplotlib:


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Creating Sample DataFrame
We'll use the following sample DataFrame for our examples:

```python
# OPTION 1:
# Sample data
data = {
    'Product': ['Smartphone', 'Tablet', 'Laptop'],
    'Price': [800, 500, 1000],
    'Quantity': [100, 30, 50],
    'Weight': [2, 3, 8]
}
# Creating DataFrame
df = pd.DataFrame(data)

# OPTION 2 (go with this for class, above is just for reference):
df = pd.read_csv('data_tech_order.csv')

print(df)
```

## Part 1: Basic Plotting with Pandas (30 minutes)

Pandas makes it easy to create basic plots using the `plot` method. The default plot type is a line plot.

#### Example 1: Line Plot


```python
# Import data from csv file
df = pd.read_csv('data_tech_order.csv')

# Line plot of Price vs Product
df.plot(x='Weight', y='Price', kind='line')
plt.title("Price vs. Weight")
plt.xlabel("Weight")
plt.ylabel("Price")
plt.show()
```

#### Example 2: Customizing Line Plot

You can customize your plots by changing colors, adding markers, and more.


```python
# Customizing line plot
df.plot(x='Product', y='Price', kind='line', color='red', linestyle='--', marker='o')
plt.title("Price of Products")
plt.xlabel("Product")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

#### Example 4: Histogram

Histograms are useful for displaying the distribution of a dataset.


```python
# Step 1: Generate sample data and create a DataFrame
# generate 1000 random numbers
data = np.random.randn(1000)
# see first 10 values of the data
data[0:10]
# create a DataFrame with the data
df = pd.DataFrame(data, columns=["Value"])
# see the first 5 rows of the DataFrame
df.head(10)

# Step 2: Save the DataFrame to a CSV file (could open this in Excel)
name_csv_file = "data_session10_example_4.csv"
df.to_csv(name_csv_file, index=False)

# Step 3: Read the data back from the CSV file
df_from_csv = pd.read_csv(name_csv_file)

# Step 4: Create a histogram from the data
#NOTICE here we are plotting from plt (not df)
plt.hist(df_from_csv["Value"], bins=30, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

**Practice Problem 4: Create Your Own Histogram**

Create a histogram for a dataset of 100 random values.

**Solution:**


```python
# Creating sample data
data = np.random.randn(100)

# Creating a DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Creating a histogram from the DataFrame
#NOTICE here we are plotting from df (not plt)
df.hist(column='Value', bins=30, edgecolor='black')

# Removing the gridlines
plt.grid(False)
#NOTICE here we are plotting from plt (not df)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

#### Example 5: Scatter Plot

Scatter plots are useful for visualizing the relationship between two variables.


```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Step 1: Create a DataFrame from the sample data
df = pd.DataFrame({'X': x, 'Y': y})

# Step 2: Save the DataFrame to a CSV file
name_csv_file = "data_session10_example5.csv"
df.to_csv(name_csv_file, index=False)

# Step 3: Read the data back from the CSV file
df_from_csv = pd.read_csv(name_csv_file)

# Step 4: Create a scatter plot from the DataFrame
plt.scatter(df_from_csv['X'], df_from_csv['Y'])
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## Part 3: Advanced Customization and Subplots (30 minutes)


#### Example 6: Creating Subplots

Subplots allow you to create multiple plots in a single figure.

```python
# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

# Step 1: Create a DataFrame from the sample data
df = pd.DataFrame({'X': x, 'Y1': y1, 'Y2': y2})

# Step 2: Save the DataFrame to a CSV file
name_csv_file = "data_session10_example6.csv"
df.to_csv(name_csv_file, index=False)

# Step 3: Read the data back from the CSV file
df_from_csv = pd.read_csv(name_csv_file)

# Step 4: Create subplots from the DataFrame
plt.subplot(2, 1, 1)  # (rows, columns, panel number)
plt.plot(df_from_csv['X'], df_from_csv['Y1'], 'r')
plt.title('First Plot')

plt.subplot(2, 1, 2)
plt.plot(df_from_csv['X'], df_from_csv['Y2'], 'g')
plt.title('Second Plot')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

**Practice Problem 6: Create Your Own Subplots**

Create a figure with two subplots:
- A line plot for the data x: [0, 1, 2, 3, 4], y: [0, 1, 4, 9, 16]
- A bar plot for the data categories: ['A', 'B', 'C'], values: [5, 10, 15]

**Solution:**


```python
# Sample data for line plot
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Sample data for bar plot
categories = ['A', 'B', 'C']
values = [5, 10, 15]

# Step 1: Create DataFrames from the sample data
df_line = pd.DataFrame({'X': x, 'Y': y})
df_bar = pd.DataFrame({'Category': categories, 'Value': values})

# Step 2: Create subplots from the DataFrames
plt.subplot(2, 1, 1)
plt.plot(df_line['X'], df_line['Y'], 'b')
plt.title('Line Plot')

plt.subplot(2, 1, 2)
plt.bar(df_bar['Category'], df_bar['Value'])
plt.title('Bar Plot')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

#### Example 7: Adding Annotations


```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Step 1: Create a DataFrame from the sample data
df = pd.DataFrame({'X': x, 'Y': y})

# Step 2: Save the DataFrame to a CSV file
name_csv_file = "data_session10_example7.csv"
df.to_csv(name_csv_file, index=False)

# Step 3: Read the data back from the CSV file
df_from_csv = pd.read_csv(name_csv_file)

# Step 4: Create the line plot from the DataFrame
plt.plot(df_from_csv['X'], df_from_csv['Y'], marker='o')

# Step 5: Add annotation
plt.annotate('Peak', xy=(5, 11), xytext=(4, 10),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Step 6: Add title and labels
plt.title("Plot with Annotation")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Step 7: Display the plot
plt.show()
```

**Practice Problem 7: Add Annotations to Your Plot**

Add an annotation to the highest point in the line plot from Practice Problem 1.

**Solution:**


```python
# Import data from csv file
df = pd.read_csv('data_tech_order.csv')

# Find the index of the row with the maximum weight
max_index = df['Weight'].idxmax()
print(max_index)

# Get the product name corresponding to the maximum quantity
max_product = df.loc[max_index, 'Product']
print(max_product)

# Get the maximum price
max_price = df.loc[max_index,'Price']
print(max_price)

# Plot the line plot
df.plot(x='Weight', y='Price', kind='line', marker='o')
plt.title("Price vs. Weight")
plt.xlabel("Weight (pounds)")
plt.ylabel("Price ($)")

print(df['Weight'][max_index])

# Add annotation for the maximum
plt.annotate('Max Price', 
             xy=(df['Weight'][max_index], max_price), 
             xytext=(df['Weight'][max_index], max_price - 90),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```
