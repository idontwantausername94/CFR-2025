## Part 1: Basic Plotting with Pandas (30 minutes)

Pandas makes it easy to create basic plots using the `plot` method. The default plot type is a line plot.

#### Example 1: Line Plot


```python
# import dataset
df = pd.read_csv("data_tech_order.csv")

# Line plot
df.plot(x='Weight', y='Price', kind='line')
plt.title("Price vs. Weight")
plt.xlabel("Weight (pounds)")
plt.ylabel("Price ($)")
```

#### Example 4: Histogram

Histograms are useful for displaying the distribution of a dataset.


```python
dir(plt)
```

```python
# Generate our data
# ************************
# STEP 1: generate 1000 random numbers
data = np.random.randn(1000)
data[:11]

# STEP 2: create DataFrame with the data
df = pd.DataFrame(data, columns=["Values"])
df.head(10)

# STEP 3: save the DataFrame to a csv file (can be used in Excel)
# without variable option
df.to_csv("data_session10_example4.csv", index=False)
# create a variable for csv file name, so that I only need to change the name in one place
name_csv_file = "data_session10_example4.csv" 
df.to_csv(name_csv_file, index=False)

# STEP 4: Bring in the data from the csv
df_from_csv = pd.read_csv(name_csv_file)
df_from_csv.head(10)

# Create plot
#***************************************
plt.hist(df_from_csv["Values"], bins=30, edgecolor='red')
plt.title("Histogram of Values")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid()
```

## Part 3: Advanced Customization and Subplots (30 minutes)


#### Example 6: Creating Subplots

Subplots allow you to create multiple plots in a single figure.

```python
# Sample Data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

# Create a DataFrame from sample data
df = pd.DataFrame({'X':x, 'Y1':y1, 'Y2':y2})
print(df)

# Create graphs
plt.subplot(3,2,1)
plt.plot(df['X'], df['Y1'], color='red')

plt.subplot(3,2,2)
plt.plot(df['X'], df['Y1'], color='orange')

plt.subplot(3,2,3)
plt.plot(df['X'], df['Y1'], color='yellow')

plt.subplot(3,2,4)
plt.plot(df['X'], df['Y1'], color='green')

plt.subplot(3,2,5)
plt.plot(df['X'], df['Y1'], color='blue')

plt.subplot(3,2,6)
plt.plot(df['X'], df['Y1'], color='purple')
```

MODIFIED IN CLASS, for the example to be:

Create a figure with eight subplots in a 4 x 2 grid that alternate with the following two datasets:
- A line plot for the data x: [0, 1, 2, 3, 4], y: [0, 1, 4, 9, 16]
- A bar plot for the data categories: ['A', 'B', 'C'], values: [5, 10, 15]

**Solution:**


```python
# Sample Data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
categories = ['A', 'B', 'C']
values = [5, 10, 15]

# Create a DataFrame from sample data
df_line = pd.DataFrame({'X':x, 'Y':y})
df_bar = pd.DataFrame({'CATEGORIES':categories, 'VALUES':values})

# Create graphs
plt.subplot(2,4,1)
plt.plot(df_line['X'], df_line['Y'])

plt.subplot(2,4,2)
plt.bar(df_bar['CATEGORIES'], df_bar['VALUES'])

plt.subplot(2,4,3)
plt.plot(df_line['X'], df_line['Y'])

plt.subplot(2,4,4)
plt.bar(df_bar['CATEGORIES'], df_bar['VALUES'])

plt.subplot(2,4,5)
plt.plot(df_line['X'], df_line['Y'])

plt.subplot(2,4,6)
plt.bar(df_bar['CATEGORIES'], df_bar['VALUES'])

plt.subplot(2,4,7)
plt.plot(df_line['X'], df_line['Y'])

plt.subplot(2,4,8)
plt.bar(df_bar['CATEGORIES'], df_bar['VALUES'])
```
